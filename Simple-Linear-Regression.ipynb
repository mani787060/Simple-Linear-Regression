```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cgpa_placement.csv')                                                                                                                
```

```python
df.head() 
```

```python
# if we plot these data in the graph

plt.scatter(df['cgpa'], df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in LPA)')
```

```python
# So, hme ab hmare data ke upr simple_LR apply krke ek model create krna hai
# So, that jb hm isko 'cgpa' denge as an i/p then, it tell about 'package' as an o/p

# So, first separate your 'rows' and 'columns'

X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
# Now, call for train test split model

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
# Now, we are going to apply our Simple_LR_Algorithm 

from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression() # we create an object for this
```

```python
lr.fit(X_train,y_train)  # "fit" ka kaam hai is model ko 'train' krna
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
# Now, we give cgpa(8.58) as an i/p from 'X_test'
# And, we look that how close our o/p comes i.e., package(4.10) as an o/p from 'y_test'

lr.predict(X_test.iloc[0].values.reshape(1,1))  # you can check for another 'cgpa' also
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.plot(X_train, lr.predict(X_train), color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in LPA)')

# So, this is the line that my LR algo. found
# that, LR algo. shows that it is "Best Fit Line" for this data
```

```python
m = lr.coef_   # this gives the value of our slope
m
```

```python
b = lr.intercept_   # this gives the value of y-intercept   
b
```

```python
# So, A/c to our equation of the line  y = mx + b

y = m*(8.58) + b
y                             # this gives the same result what I get on by predicting 'iloc[0].values'  i.e., 3.89 
```

```python
# if your 'cgpa' is not in the data, despite you can calculate it's 'package'          

y = m*(9.5) + b      # we take an example that (cgpa=4.5) as an i/p
y                    # And, get an o/p by appplynig this formula
```
